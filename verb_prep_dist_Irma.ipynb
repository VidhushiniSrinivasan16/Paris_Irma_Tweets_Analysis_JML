```python
def NounPrepDist(df,lang):
    filtered_df = df.loc[df['lang'] == lang]
    print(len(filtered_df['text']))
    freqs = defaultdict(int)
    for tweet in filtered_df['text']:
        processed_text = nlp(tweet)        
        for w in processed_text:
            if (w.pos_ == "PART" or w.pos_ == "ADP") and w.head.pos_ == "VERB":
                freqs[w.head.lemma_,w.lemma_]+=1
    return freqs
```

```python
def VerbPrepEntropy(dist):
     pairs = list(dist.keys())
     verbH = defaultdict(float) 
     verbs,_ = zip(*pairs)
     verbs = set(verbs)     
     for v in verbs:
             counts = [dist[p] for p in pairs if p[0]==v]
             verbH[v] = EntropyEst(counts)/log(2.)
     return verbH
```

```python
def EntropyEst(freqs):
     n = float(sum(freqs))
     f1 = float(len([x for x in freqs if x==1]))
     f2 = float(len([x for x in freqs if x==2]))
     if f2>0:
         A = 2. * f2 / ((n - 1.)*f1 + 2. * f2)
     elif f1>0:
         A = 2. / ((n - 1.)*(f1 - 1.) + 2.)
     else:
         A = 1.
     v = array(freqs)
     v = v[v<n]
     R = sum([_CWJ_aux(x,n) for x in v])
     r = arange(1,n)
     if A != 1.0:
         R -= f1*(log(A)+sum(((1.-A)**r)/r))*((1.-A)**(1.-n))/n
     return R
```

```python
_CWJ_Chart = {}
def _CWJ_aux(Xi,n):
     if not (Xi,n) in _CWJ_Chart:
         _CWJ_Chart[Xi,n] = Xi * sum(1./arange(Xi,n))/n
     return _CWJ_Chart[Xi,n]
```
